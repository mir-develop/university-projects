Если пропуски есть - заполнить их пустой строкой для строк, нулем для чисел, 31 декабря 2016 года для даты

Построить график зависимости сытости от цены (Hunger от Cost)

```python
plt.figure(figsize=(5,5))
x = df["Hunger"]
y = df["Cost"]

plt.scatter(x, y, c="b",edgecolor='r', alpha=1)
            
plt.xlabel("Hunger")
plt.ylabel("Cost")
plt.show()
```

Построить гистограмму распределения для атрибутов Filling и Meat на соседних графиках (subplot)

```python
fig, (ax1) = plt.subplots(1, sharey=True)
x1 = df["Fillings"] 
ax1.plot(x1, 'r', alpha=1)
ax1.set(title='Fillings') 


fig, (ax1) = plt.subplots(1, sharey=True)
x1 = df["Meat"] 
ax1.plot(x1, 'y', alpha=0.8) 
ax1.set(title='Meat') 

plt.show()
```

Построить график зависимости цены от даты (Cost от Date)

```python
plt.figure(figsize=(8,30))

x = df["Cost"]
y = df["Date"]

plt.plot(x, y, c="m", alpha=1)

            
plt.xlabel("Cost")
plt.ylabel("Date")
plt.show()
```
